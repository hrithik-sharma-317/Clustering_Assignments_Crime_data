# Hierarchical Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt 
import seaborn as sns
```

```python
crime=pd.read_csv("D:\\Data_science_Assignments\\Clustering\\crime_data.csv")
crime
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data_norm = pd.DataFrame(trans.fit_transform(crime.iloc[:,1:]))
data_norm
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
Single= linkage(data_norm, method="single",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(Single) 
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
Avg= linkage(data_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(Avg) 
plt.show() 
```

```python
from sklearn.cluster import AgglomerativeClustering 
Avg = AgglomerativeClustering(n_clusters=5, linkage='average',affinity = "euclidean").fit(data_norm) 
cluster_labels=pd.Series(Avg.labels_)
cluster_labels
crime['Clust']=cluster_labels 
crime 
```

```python
crime.iloc[:,1:].groupby(crime.Clust).mean()
```

```python
data = crime[(crime.Clust==0)]
data
```

```python
data = crime[(crime.Clust==3)]
data
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans

fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(data_norm)
    WCSS.append(clf.inertia_) 
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show() 
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(data_norm) 
```

```python
clf.labels_
```

```python
y_kmeans
```

```python
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
md=pd.Series(y_kmeans)  
crime['clust']=md
crime
```

```python
crime.iloc[:,1:].groupby(crime.clust).mean()
```

```python
crime.plot(x= "Assault",y ="Murder",c=clf.labels_,kind="scatter",s=30 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```

# DB Scan Clustering

```python
from sklearn.cluster import DBSCAN
array=crime.iloc[:,1:].values
array 
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X 
```

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=8)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
dbscan = DBSCAN(eps=1.45, min_samples=4)
dbscan.fit(X)  
```

```python
dbscan.labels_
```

```python
d1 = dbscan.labels_
d1
```

```python
import sklearn
sklearn.metrics.silhouette_score(X,d1)
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(X) 
```

```python
y_kmeans
```

```python
sklearn.metrics.silhouette_score(X, y_kmeans)
```
